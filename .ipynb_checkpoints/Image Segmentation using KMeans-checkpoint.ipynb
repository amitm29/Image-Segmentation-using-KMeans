# Image Segmentation using K-Means Clustering Algorithm

```python
import matplotlib.pyplot as plt
import cv2
```

### Reading the image

```python
img = cv2.imread('elephant.jpg')

#converting BGR image to RGB (as openCV reads the image in BGR)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
#Plotting the image
plt.imshow(img)
plt.axis("off")
plt.show()
```

### Preparing the image data for KMeans

```python
import numpy as np
```

```python
#Flattening the image
img = np.array(img)
original_shape = img.shape
print(original_shape)
image_data = np.reshape(img,(-1,3))
print(image_data.shape)
```

Now we have 165000 examples (i.e. pixels in this case) with 3 features (1 for each channel - R, G, and B) in each example.

## Dominant color extraction using K-means clustering algorithm

### Training the KMeans Clustering Algorithm

Traing the KMeans Algorithm on the given image will give us the k dominant colors from the given image by making k clusters from the given pixels of the image using k Means Clustering algorithm.

```python
from sklearn.cluster import KMeans
```

```python
k = 5       #number of dominant colors to detect
```

```python
km = KMeans(n_clusters=k)
```

```python
km.fit(image_data)
```

```python
colors = km.cluster_centers_
```

Cluster centers gives us the dominant colors in the image.

```python
colors
```

```python
colors = np.array(colors, dtype='uint8')  #converting colors into 8-bit unsigned integers i.e. 0-255
```

```python
colors
```

### Creating Color Swatches

```python
# Plotting the color swatches for k dominant colors detected by K-Means algorithm
plt.figure(figsize=(3,10))
i = 1
for color in colors:
    plt.subplot(1,k,i)
    plt.axis("off")
    
    i+=1
    
    a = np.zeros((100,100,3), dtype='uint8')
    a[:,:,:] = color
    plt.imshow(a)
```

## Segmenting our original image

```python
new_img = np.zeros(image_data.shape,dtype='uint8')
```

```python
for ix in range(image_data.shape[0]):
    new_img[ix] = colors[km.labels_[ix]]
```

```python
new_img = np.reshape(new_img, original_shape)
```

```python
plt.imshow(new_img)
plt.axis('off')
plt.show()
```
